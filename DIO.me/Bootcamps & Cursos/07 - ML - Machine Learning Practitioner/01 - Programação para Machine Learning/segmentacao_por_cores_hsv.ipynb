```python
bgr_img = cv.imread('./image.jpg')

hsv_img = cv.cvtColor(bgr_img, cv.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1), plt.imshow(bgr_img, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 2), plt.imshow(hsv_img, cmap='gray')
plt.title('HSV Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
rgb_green = np.uint8([[[0, 255, 0]]])
hsv_green = cv.cvtColor(rgb_green, cv.COLOR_RGB2HSV)[0, 0, :]
print(hsv_green)
```

```python
hsv_img = cv.cvtColor(bgr_img, cv.COLOR_BGR2HSV)

lower_green = np.array([36, 25, 25])
upper_green = np.array([86, 255, 255])

mask = cv.inRange(hsv_img, lower_green, upper_green)

lower_red = np.array([161, 155, 84])
upper_red = np.array([179, 255, 255])

mask2 = cv.inRange(hsv_img, lower_red, upper_red)

lower_blue = np.array([94, 80, 2])
upper_blue = np.array([126, 255, 255])

mask3 = cv.inRange(hsv_img, lower_blue, upper_blue)

plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1), plt.imshow(mask, cmap='gray')
plt.title('Mask'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 2), plt.imshow(mask2, cmap='gray')
plt.title('Mask2'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 3), plt.imshow(mask3, cmap='gray')
plt.title('Mask3'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
rgb_res = cv.bitwise_and(bgr_img, bgr_img, mask=mask)

plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1), plt.imshow(rgb_res, cmap='gray')
```
