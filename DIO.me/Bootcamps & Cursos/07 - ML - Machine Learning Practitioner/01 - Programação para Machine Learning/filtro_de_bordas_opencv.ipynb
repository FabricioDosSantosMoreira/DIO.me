```python
import cv2 as cv
import numpy as np

from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
!wget -q https://cdn.esahubble.org/archives/images/screen/opo9941a.jpg -O './image.jpg'
```

```python
img_path = './image.jpg'

img = cv.imread(img_path)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

gaussian_img = cv.GaussianBlur(gray, (3, 3), 0)
```

```python
cv2_imshow(gaussian_img)
```

```python
sobelx = cv.Sobel(gaussian_img, cv.CV_64F, 1, 0, ksize=5)
sobely = cv.Sobel(gaussian_img, cv.CV_64F, 0, 1, ksize=5)
sobelxy = cv.Sobel(gaussian_img, cv.CV_64F, 1, 1, ksize=5)

plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1), plt.imshow(img, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 2), plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 3), plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 4), plt.imshow(sobelxy, cmap='gray')
plt.title('Sobel XY'), plt.xticks([]), plt.yticks([])

plt.show()
```
